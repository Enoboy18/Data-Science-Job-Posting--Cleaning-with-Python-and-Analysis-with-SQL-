```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Glass_Door_Data_Science_Cleaned_Data.csv")
```

```python
pd.set_option("display.max_rows", 672)
df
```

```python
df.drop(columns="Competitors", inplace=True)
df.drop(columns="Job Description", inplace=True)
```

```python
df
```

```python
df.replace([-1, '-1', 'Unknown'], np.nan, inplace=True)
df
```

```python
df['Rating'] = df['Rating'].astype(float)
df['Founded'].fillna(0, inplace=True)
df['Company Age'].fillna(0, inplace=True)
df['Founded'] = df['Founded'].astype(int)
df['Company Age'] = df['Company Age'].astype(int)
df
```

```python
df.to_csv('output_file.csv', index=False)
```
